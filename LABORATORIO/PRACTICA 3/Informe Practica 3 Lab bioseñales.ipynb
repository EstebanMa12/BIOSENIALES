```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob
import scipy.io as sio
from scipy.stats import shapiro, levene, mannwhitneyu
```

## Importando archivos

```python
archivos_control = glob.glob('./control/*.mat') # retorna una lista con las rutas de todos los archivos en la carpeta control que terminen en .mat
archivos_parkinson = glob.glob('./parkinson/*.mat')# retorna una lista con las rutas de todos los archivos en la carpeta parkinson que terminen en .mat
print('Hay ' + str(len(archivos_control)) + ' sujetos de control')
print('Hay ' + str(len(archivos_parkinson)) + ' sujetos de parkinson')
```

## 1. Implemente una función que reciba una señal de múltiples canales y épocas y calcule la Energía de promedio de cada canal.
Función para calcular energía

```python
def calcMeanEnergy(data):#Función que permite calcular el promedio de la energía de cada canal
    energy=np.sum(np.square(np.abs(data)), axis=1) #Cálculo de la energía de cada canal
    energy_avg=np.mean(energy, axis=1) #Promedio de la energía de cada canal
    return energy_avg
```

## 2. Calcule la energía de cada canal promediada por épocas para cada sujeto, esto para ambos grupos poblacionales. Guarde esta información en un DataFrame de columnas ‘canal’ y filas ‘#sujeto’ con los valores de energía calculados, un DataFrame para cada grupo poblacional.

### Carga de archivos de control, cálculo de energía promedio y generación de dataframe

```python
energy_control = []#arreglo vacio en donde se guardan los promedios de energia de cada canal 
for archivo in archivos_control:# Para cada archivo presente en archivos_control
    mat_contents = sio.loadmat(archivo) # Carga el archivo .mat
    data_original = mat_contents['data']; #Carga de los datos
    data = data_original.copy() # Se realiza una copia de los datos originales para trabajar sobre ellos
    energy_mean = calcMeanEnergy(data) #Calculo del promedio de la energía 
    energy_control.append(energy_mean) #Insertar en el arreglo vacio el resultado para cada archivo
df_control=pd.DataFrame(energy_control) # Creación del dataframe 
df_control.columns.names = ['Canal'] #Nombre de las columnas
df_control.index = np.arange(1, len(df_control)+1) # Indice de las filas 
df_control.index.names = ['#sujeto'] # nombre de las filas
df_control
```

### Carga de archivos de parkinson, cálculo de energía promedio y generación de dataframe

```python
energy_parkinson = []
for archivo in archivos_parkinson: # Se realiza el mismo procedimiento que con los datos de control para la creación del dataframe
    mat_contents = sio.loadmat(archivo)
    data_original = mat_contents['data'];
    data = data_original.copy()
    energy_mean = calcMeanEnergy(data)
    energy_parkinson.append(energy_mean)
df_parkinson=pd.DataFrame(energy_parkinson)
df_parkinson.columns.names = ['Canal']
df_parkinson.index = np.arange(1, len(df_parkinson)+1)
df_parkinson.index.names = ['#sujeto']
df_parkinson
```

## 3. Determine si existe diferencia estadística entre canales de cada grupo de sujetos a través de una prueba t. Compruebe los supuestos necesarios para realizar una prueba t, esto es: Normalidad de la variable, independencia (se asume que los grupos son independientes), y homocedasticidad (use una prueba de Levene), finalmente realice la prueba t para determinar si existen diferencias entre los canales entre grupos de sujetos. De no cumplirse los requisitos, realice entonces un análisis no paramétrico (prueba U de Mann-Whitney). Este numeral tiene como objetivo identificar los canales que entregan información diferencial entre pacientes Sanos y con enfermedad de Parkinson. 

### Pruebas de normalidad
Para las pruebas de normalidad haremos uso de la prueba Shapiro-Wilk. 
Se establece:
* Hipótesis nula (H0): La distribución es normal
* Hipótesis alternativa (H1): La distribución no es normal
* Nivel de significancia de 0.05

```python
alpha_shapiro_ct = 0.05
for col in df_control.columns: #Para cada columna presente en el dataframe de control
    stat, p = shapiro(list(df_control[col])) #Implementación de la prueba Shapiro-Wilk
    print(f'Grupo de control - Canal {col}: Estadístico de prueba={stat:.3f}, p={p:.3f}')
    if p > alpha_shapiro_ct:
        print(f'Los datos del Canal {col} del grupo de control parecen seguir una distribución normal (no se puede rechazar la hipótesis nula)\n')
    else:
        print(f'Los datos del Canal {col} del grupo de control no parecen seguir una distribución normal (se rechaza la hipótesis nula)\n')
```

```python
alpha_shapiro_pk = 0.05
for col in df_parkinson.columns:#Para cada columna presente en el dataframe de parkinson
    stat, p = shapiro(list(df_parkinson[col]))#Implementación de la prueba Shapiro-Wilk
    print(f'Grupo de parkinson - Canal {col}: Estadístico de prueba={stat:.3f}, p={p:.3f}')
    if p > alpha_shapiro_pk:
        print(f'Los datos del canal {col} del grupo de parkinson parecen seguir una distribución normal (no se puede rechazar la hipótesis nula)\n')
    else:
        print(f'Los datos del canal {col} del grupo de parkinson no parecen seguir una distribución normal (se rechaza la hipótesis nula)\n')
```

### Pruebas de homocedasticidad
Para las pruebas de homocedasticidad haremos uso de la prueba de Levene. 
Se establece:
* Hipótesis nula (H0): La varianza poblacional es constante
* Hipótesis alternativa (H1): La varianza poblacional no es constante
* Nivel de significancia de 0.05

```python
alpha_levene = 0.05
for col in df_control.columns:
    stat, p = levene(list(df_control[col]), list(df_parkinson[col]))#Implementación de la prueba levene
    print(f'Canal {col}: Estadístico de prueba={stat:.3f}, p={p:.3f}')
    if p > alpha_levene:
        print(f'Las varianzas de los canales {col} parecen ser constantes entre los grupos (no se puede rechazar la hipótesis nula)\n')
    else:
        print(f'Las varianzas de los canales {col} no parecen ser constantes entre los grupos (se rechaza la hipótesis nula)\n')
```

### Resultados pruebas de normalidad y homocedasticidad
Luego de realizar las pruebas para comprobar los supuestos de normalidad y homocedasticidad encontramos que los datos no están distribuidos de forma normal en ningún canal (no cumplen con la prueba de normalidad) y que todos los pares de canales tienen una varianza constante (cumplen con la prueba de homocedasticidad).

Al no cumplirse todos los supuestos, se optó por realizar una prueba U de Mann-Whitney.

### Prueba U de Mann-Whitney
Para la prueba U de Mann-whitney se establece:
* Hipótesis nula (H0): No existe diferencia estadística entre canales de cada grupo de sujetos.
* Hipótesis alternativa (H1): Existe diferencia estadística entre canales de cada grupo de sujetos.
* Nivel de significancia de 0.05

```python
alpha_u = 0.05
for col in df_control.columns:
    stat, p = mannwhitneyu(list(df_control[col]), list(df_parkinson[col]), alternative='two-sided')#Implementación de la prueba U de Mann-Whitney
    print(f'Canal {col}: Estadístico U={stat:.3f}, p={p:.3f}')
    if p > alpha_u:
        print(f'No se puede rechazar la hipótesis nula de que no existe diferencia estadística entre los datos del canal {col} de ambos grupos.\n')
    else:
        print(f'Se rechaza la hipótesis nula de que existe diferencia estadística entre los datos del canal {col} de ambos grupos.\n')
```

## Análisis



A pesar de que la prueba de normalidad no se cumplió desde un inicio, se realizó la prueba de homocedasticidad para cumplir con los items planteados

De acuerdo a los resultados obtenidos por la prueba de U de Mann Whitney, se afirma que no existe una diferencia estadística significativa entre el promedio de la energía de cada canal de los grupos poblacionales, como lo son el de control y el de parkinson, esto se debe a que al realizar la prueba mencionada para cada canal, se aceptan todas las hipótesis nulas, lo que sugiere que las muestras no son significativamente diferentes entre sí, en términos de la energía promedio.

* Con el resultado obtenido en la prueba de U de Mann whitney se puede concluir que no es posible establecer los canales que entregan información diferencial entre pacientes Sanos y con enfermedad de Parkinson, puesto a que estos determinan  que estadísticamente las muestras son iguales. Por tanto, se procedió a realizar el análisis Mann-Whitney.
